# 🎭 I Transformer: "Non conta solo con chi vai, ma che ruolo hai nella storia"

Se con Word2Vec abbiamo imparato che una parola è definita dai suoi vicini ("Dimmi con chi vai..."), con i **Transformer** facciamo un salto rivoluzionario. Passiamo dai vettori di singole parole ai **Sentence Embeddings**: vettori che rappresentano il significato di un'intera frase.

L'idea di base non è più solo guardare i vicini, ma chiedersi: **"Cosa stai facendo esattamente in questa specifica scena?"**

---

## 1. Perché Word2Vec non basta più
Word2Vec ha un grande limite: è **statico**. Per Word2Vec, la parola "capo" ha sempre lo stesso vettore (lo stesso "punto" nello spazio), indipendentemente dal contesto. 

Il **Transformer** invece è come un camaleonte: cambia il "colore" (il vettore) di ogni parola in base a tutto ciò che la circonda. 

> **Esempio:**
> 1. "Il **capo** ha firmato il contratto." (Significa: Leader/Manager)
> 2. "Ho comprato un **capo** firmato." (Significa: Vestito)
>
> In Word2Vec, "capo" è un unico punto medio confuso. Nel Transformer, la parola "capo" genera due vettori completamente diversi perché **capisce il ruolo della parola nella frase**.

---

## 2. Il Segreto: L'Attenzione
Immaginate di leggere una frase complessa. I vostri occhi non guardano ogni parola con la stessa intensità. Se leggete: 
*"Il gatto, dopo aver mangiato ed essere uscito in giardino, è tornato perché aveva **fame**"*.

Quando arrivate alla parola **"fame"**, il vostro cervello accende un riflettore (un'attenzione) sulla parola **"gatto"**. Non vi importa di "giardino" o "uscito", vi interessa capire *chi* ha fame.

Il Transformer fa esattamente questo: usa dei **"Riflettori di Attenzione"** per capire quali parole sono collegate tra loro, anche se sono molto lontane all'interno della frase.



---

## 3. Da Parole a Vettore Unico: Il Sentence Embedding
Mentre con Word2Vec dovevamo fare la "media" (il frullatore) dei vettori per provare a capire una frase, i Transformer creano un **Sentence Embedding** (un vettore della frase) in modo intelligente:

* **Word2Vec (La Media):** Mescola le parole come ingredienti di un'insalata. Il sapore è un mix, ma perdiamo l'ordine e la struttura.
* **Transformer (La Sintesi):** Legge la frase come uno chef esperto. Capisce come le parole interagiscono tra loro e crea un "sapore unico" (un unico vettore di numeri) che rappresenta il **senso profondo** dell'intero messaggio.

---

## 4. L'Architettura: Sotto il Cofano dei Giganti
Tutti i modelli moderni che sentiamo nominare oggi (BERT, Gemini, Claude, ChatGPT) si basano sull'architettura Transformer. 

Questa struttura permette al computer di leggere tutte le parole di una frase **contemporaneamente** (in parallelo), invece di una dopo l'altra come facevano i vecchi modelli. Questo gli permette di cogliere relazioni che prima erano impossibili da vedere.

## 5. Base vs Fine-tuning: Dal "Tuttologo" allo "Specialista"

I modelli Transformer attraversano un percorso di crescita diviso in due grandi momenti. Immaginiamoli come il percorso di studi di una persona.



---

### 🏛️ Il Modello Base (Pre-training)
È il risultato della prima fase di addestramento su enormi quantità di dati generici.

* **Cosa fa:** Presa una frase, ti restituisce la sua **Rappresentazione** (il Sentence Embedding).
* **Il Ruolo:** È un **"Fornitore di Dati"**. Ti consegna una mappa numerica del significato della frase.
* **Come si usa:**
    * La usi come "carburante" per addestrare un classificatore esterno (es. una Regressione Logistica o una SVM).
    * La usi per i **motori di ricerca semantici**: confronti i vettori per trovare testi che hanno significati simili.
* **In breve:** "Ti do il DNA del testo, ora decidi tu come usarlo."

---

### 🎯 Il Modello Specializzato (Fine-tuning)
È il risultato di un secondo addestramento mirato su un compito specifico (es. analisi del sentiment, spam detection, summarization).

* **Cosa fa:** Non ti dà più una rappresentazione astratta, ma ti fornisce direttamente l'**Output** finale relativo al task.
* **Il Ruolo:** È uno **"Specialista"**. Usa la sua conoscenza interna per risolvere il problema senza passarti dati intermedi.
* **Come si usa:** Lo integri direttamente nel tuo flusso di lavoro. Non serve un classificatore esterno: il modello **è** il classificatore.
* **In breve:** "Non chiedermi i vettori; dimmi cosa vuoi che faccia e io lo eseguo."



---

### 💡 

| Caratteristica | Modello Base (Tuttologo) | Modello Fine-tuned (Specialista) |
| :--- | :--- | :--- |
| **Cosa sputa fuori?** | Rappresentazioni (vettori) | Risultato finale (Task specifico) |
| **Come si usa?** | Come base per altri strumenti | Come strumento pronto all'uso |
| **Conoscenza** | Vasta ma generica | Profonda e specifica |
| **Analoga** | Laureato in Lettere | Critico Letterario esperto di Dante |

> La grande rivoluzione dei Transformer è proprio questa: non devi più costruire un modello da zero. Prendi il "cervello" di un modello base (che sa già come funziona la lingua) e gli insegni un mestiere specifico (Fine-tuning) in pochissimo tempo e con pochi esempi.

---

### 6. Addio Pre-processing: Il Tokenizzatore Integrato

Con i Transformer, il "lavoro sporco" di pulizia manuale del testo (rimozione punteggiatura, minuscole, stemming) è diventato obsoleto.

* **Il "Butler" integrato:** Ogni Transformer ha il suo **Tokenizzatore dedicato**. È parte integrante del modello: non dobbiamo più pulire il testo a mano.
* **Tokenizzazione "Intelligente":** I tokenizzatori moderni usano tecniche come le *Sub-word tokenization* (es. WordPiece o BPE). Se incontrano una parola rara, non la scartano, ma la spezzano in pezzi più piccoli (es. "anti"+"dis"+"establish"), riuscendo a dare un senso anche a termini mai visti prima.

---

## 7. Comparazione

| Caratteristica | Word2Vec (Dizionario) | Transformer (Lettore) |
| :--- | :--- | :--- |
| **Tipo di Vettore** | Statico (sempre uguale) | Dinamico (cambia col contesto) |
| **Memoria** | Breve (guarda solo i vicini) | Globale (guarda tutta la frase) |
| **Punto di forza** | Veloce e semplice | Capisce il senso compiuto |
| **Output Finale** | Lista di parole | **Vettore unico della frase (Sentence Embedding)** |

> Se Word2Vec ci ha insegnato a tradurre le parole in numeri, i Transformer ci hanno insegnato a tradurre i **pensieri** e i **contesti** in numeri. È questo che permette ai computer di capire il sarcasmo, le negazioni e le sfumature di una conversazione.

# Parte pratica

```python
import pandas as pd
from datasets import load_dataset
from sklearn.model_selection import train_test_split
from sentence_transformers import SentenceTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from transformers import pipeline

# Scarichiamo dati per i task
imdb = load_dataset("imdb", split="train[:10000]") # Classificazione binaria
fin_sentiment = load_dataset("financial_phrasebank", "sentences_allagree", split="train[:500]", trust_remote_code=True)
news_data = load_dataset("ag_news", split="test[:200]") # Motore di ricerca
ner_data = load_dataset("conll2003", split="test[:50]") # NER
```

## 1. Otteniamo gli embedding e addestriamo un classificatore

```python
imdb.head()
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2') # Scarichiamo il modello pre-addestrato
X = model.encode(imdb['text']) # creiamo i vettori
y = imdb['label'] # isoliamo le label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # Suddividiamo i dati in training e test

clf = LogisticRegression() # Creiamo un classificatore semplice
clf.fit(X_train, y_train) # Addestriamo il classificatore
preds = clf.predict(X_test) # Facciamo le predizioni
print(f"Accuratezza Classificazione: {accuracy_score(y_test, preds):.2f}")
```

## 2. Creiamo un motore di ricerca semantico

```python
news_data.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

news_embeddings = model.encode(news_data['text']) #Creiamo i vettori per le notizie

def cerca_news(query, top_k=3):
    q_emb = model.encode([query]) # creiamo il vettore per la query
    scores = cosine_similarity(q_emb, news_embeddings) # calcoliamo la similarità tra la query e le notizie
    idx = scores.argsort()[0][-top_k:][::-1] # ordiniamo le notizie in base alla similarità e prendiamo le top_k
    return [news_data['text'][i] for i in idx] # Funzione per cercare le notizie più rilevanti in base alla query

print(cerca_news("economic growth in europe"))
```

## 3. Usiamo un modello fine tunato per fare sentiment direttamente

```python
fin_sentiment.head()
```

```python
fin_pipe = pipeline("sentiment-analysis", model="ProsusAI/finbert") #pipeline ci permette di selezionare un task ed un modello per quel task

# Mapping label per confrontarle con quelle del dataset (0:negative, 1:neutral, 2:positive)
labels_map = {'negative': 0, 'neutral': 1, 'positive': 2}
texts = fin_sentiment['sentence'] # isoliamo i testi da analizzare
y_true = fin_sentiment['label'] # isoliamo le label vere (0, 1, 2) per calcolare l'accuratezza

results = fin_pipe(texts) # applichiamo la pipeline inizializzata
y_pred = [labels_map[r['label']] for r in results] # applichiamo il mapping per ottenere le label predette (0, 1, 2)

print(f"Accuratezza Sentiment Finanziario: {accuracy_score(y_true, y_pred):.2f}") #calcoliamo l'accuratezza 
```

## 4. Named Entity Recognition

```python
ner_data.head()
```

```python
ner_pipe = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english", aggregation_strategy="simple") # aggregation_strategy ci permette di unire le parole che fanno parte della stessa entità

df_ner = pd.DataFrame({'text': ner_data['tokens']})
df_ner['text'] = df_ner['text'].apply(lambda x: " ".join(x)) # Uniamo i token

# Applichiamo NER e salviamo in colonna
df_ner['entities'] = df_ner['text'].apply(lambda x: [ent['word'] for ent in ner_pipe(x)])

print(df_ner.head())
```
