## 1.2 L'obiettivo: rappresentare il testo

### La sfida centrale di NLP

Il compito principale dell'NLP è trovare una **rappresentazione numerica efficace** di parole, frasi e testi.

Perché?
1. I computer capiscono i **numeri**, non le parole
2. Una volta che il testo è numerico, possiamo applicare algoritmi di Machine Learning

### Confronto: Testo grezzo vs. Rappresentazione numerica

```python
import numpy as np

# Testo grezzo (quello che gli umani leggono)
sentence = "I love machine learning"
print("=== TESTO GREZZO ===")
print(f"Frase: {sentence}")
print(f"Tipo: {type(sentence)}")
print()

# Rappresentazione 1: Semplice - codificare ogni parola con un numero
print("=== RAPPRESENTAZIONE 1: One-Hot Encoding (Simbolico) ===")
vocab = {"I": 0, "love": 1, "machine": 2, "learning": 3}
words = sentence.split()
encoded = [vocab[word] for word in words]
print(f"Parole: {words}")
print(f"Codificate: {encoded}")
print(f"Tipo: lista di interi")
print()

# Rappresentazione 2: Bag of Words (contare le parole)
print("=== RAPPRESENTAZIONE 2: Bag of Words (Statistica) ===")
from collections import Counter
word_counts = Counter(words)
bow_vector = [word_counts.get(word, 0) for word in sorted(vocab.keys())]
print(f"Vocabolario ordinato: {sorted(vocab.keys())}")
print(f"Bag of Words: {bow_vector}")
print(f"Significato: [1 volta 'I', 1 volta 'love', 1 volta 'machine', 1 volta 'learning']")
print()

# Rappresentazione 3: Embedding (vettore denso, simula i significati)
print("=== RAPPRESENTAZIONE 3: Word Embeddings (Deep Learning) ===")
np.random.seed(42)
# In realtà gli embeddings si calcolano da dati enormi, qui usiamo numeri casuali per illustrare
embeddings = {
    "I": np.array([0.2, -0.5, 0.8]),
    "love": np.array([0.9, 0.3, -0.1]),
    "machine": np.array([0.1, 0.7, 0.4]),
    "learning": np.array([0.3, 0.6, 0.5])
}
print(f"Ogni parola è un vettore denso (3 dimensioni in questo esempio):")
for word in words:
    print(f"  {word}: {embeddings[word]}")
print()
# Calcola il vettore medio della frase
sentence_embedding = np.mean([embeddings[word] for word in words], axis=0)
print(f"Embedding della FRASE (media): {sentence_embedding}")
print(f"Idea: parole simili hanno vettori simili")
```

### 1.3.1 Text Classification (Classificazione di testi)

**Definizione**: Assegnare a un testo una o più categorie predefinite.

**Esempi**:
- **Spam detection**: Un'email è spam o no?
- **Sentiment analysis**: Una recensione è positiva, negativa o neutra?
- **Topic classification**: Un articolo parla di sport, politica, tecnologia o economia?

```python
# --- TASK 1: Text Classification (Sentiment Analysis) ---

def simple_sentiment_classifier(text):
    """
    Classificatore basato su dizionario (Symbolic approach).
    Approccio naïve: conta le parole positive e negative.
    """
    positive_words = ['good', 'great', 'excellent', 'wonderful', 'amazing', 'love', 'beautiful']
    negative_words = ['bad', 'terrible', 'awful', 'horrible', 'hate', 'ugly', 'disgusting']
    
    text_lower = text.lower() # perchè facciamo questo?
    positive_count = sum(1 for word in positive_words if word in text_lower)
    negative_count = sum(1 for word in negative_words if word in text_lower)
    
    if positive_count > negative_count:
        return "POSITIVE", positive_count - negative_count
    elif negative_count > positive_count:
        return "NEGATIVE", negative_count - positive_count
    else:
        return "NEUTRAL", 0

# Test
reviews = [
    "This movie is absolutely wonderful and amazing!",
    "Terrible and awful experience, I hate it.",
    "The weather is nice today."
]

print("=== TASK 1: Text Classification (Sentiment Analysis) ===")
for review in reviews:
    sentiment, score = simple_sentiment_classifier(review)
    print(f"Testo: '{review}'")
    print(f"Sentimento: {sentiment} (score: {score})")
```

### 1.3.2 Named Entity Recognition (NER)

**Definizione**: Identificare e classificare entità nominate (persone, luoghi, organizzazioni, ecc.) nel testo.

**Esempi**:
- Trovare nomi di persone: "Mario Rossi"
- Trovare nomi di città: "Milano", "Roma"
- Trovare nomi di aziende: "Apple", "Microsoft"

```python
import re

# --- TASK 2: Named Entity Recognition (NER) ---

def simple_ner(text):
    """
    Riconoscimento semplice di entità basato su regole e dizionari.
    Approccio naïve: usa una lista di entità conosciute.
    """
    # Dizionari di entità conosciute
    people = ['Mario', 'Luigi', 'John', 'Alice']
    cities = ['Milano', 'Roma', 'Venice', 'New York']
    companies = ['Apple', 'Google', 'Microsoft', 'Amazon']
    
    entities = {}
    
    # Cerca persone
    for person in people:
        if person in text:
            entities[person] = 'PERSON'
    
    # Cerca città
    for city in cities:
        if city in text:
            entities[city] = 'LOCATION'
    
    # Cerca aziende
    for company in companies:
        if company in text:
            entities[company] = 'ORGANIZATION'
    
    return entities

# Test
texts = [
    "Mario lives in Milano and works for Google.",
    "Alice visited Rome and had lunch at an Apple store."
]

print("=== TASK 2: Named Entity Recognition (NER) ===")
for text in texts:
    entities = simple_ner(text)
    print(f"Testo: '{text}'")
    print(f"Entità trovate: {entities}")
    print()
```

### 1.3.3 Part-of-Speech (POS) Tagging

**Definizione**: Assegnare a ogni parola una categoria grammaticale (sostantivo, verbo, aggettivo, ecc.).

**Esempi**:
- "The cat is sleeping" → The (DET), cat (NOUN), is (VERB), sleeping (VERB)

```python
# --- TASK 3: Part-of-Speech Tagging ---

def simple_pos_tagger(sentence):
    """
    Etichettatura semplice basata su dizionario.
    Approccio naïve: assegna la categoria più comune.
    """
    # Dizionario parola -> categoria
    pos_dict = {
        'the': 'DET',
        'a': 'DET',
        'cat': 'NOUN',
        'dog': 'NOUN',
        'is': 'VERB',
        'runs': 'VERB',
        'beautiful': 'ADJ',
        'happy': 'ADJ',
        'quickly': 'ADV',
        'slowly': 'ADV'
    }
    
    words = sentence.lower().split()
    tagged = []
    
    for word in words:
        # Rimuovi punteggiatura
        clean_word = word.strip('.,!?;:')
        tag = pos_dict.get(clean_word, 'UNKNOWN')
        tagged.append((word, tag))
    
    return tagged

# Test
sentences = [
    "The cat is beautiful.",
    "A dog runs quickly."
]

print("=== TASK 3: Part-of-Speech Tagging ===")
for sentence in sentences:
    tagged = simple_pos_tagger(sentence)
    print(f"Frase: '{sentence}'")
    print("POS tags:")
    for word, tag in tagged:
        print(f"  {word}: {tag}")
    print()
```

### 1.3.4 Information Extraction (IE)

**Definizione**: Estrarre informazioni strutturate (relazioni, fatti) da testi non strutturati.

**Esempio**: Da un contratto, estrarre: data_firma, partecipanti, importo, condizioni.

```python
# --- TASK 4: Information Extraction ---

def simple_information_extraction(text):
    """
    Estrazione di informazioni basata su pattern e regex.
    Approccio naïve: cerca parole chiave e pattern specifici.
    """
    extracted_info = {}
    
    # Cerca importo (numero seguito da $, €, ecc.)
    amount_match = re.search(r'\$(\d+)', text)
    if amount_match:
        extracted_info['amount'] = amount_match.group(1)
    
    # Cerca date (formato DD/MM/YYYY)
    date_match = re.search(r'(\d{1,2}/\d{1,2}/\d{4})', text)
    if date_match:
        extracted_info['date'] = date_match.group(1)
    
    # Cerca nomi (parole capitalizzate)
    names = re.findall(r'\b([A-Z][a-z]+)\b', text)
    if names:
        extracted_info['names'] = names
    
    return extracted_info

# Test
contract = "Agreement between Mario and Alice on 15/03/2024 for $5000."

print("=== TASK 4: Information Extraction ===")
print(f"Testo: '{contract}'")
info = simple_information_extraction(contract)
print(f"Informazioni estratte:")
for key, value in info.items():
    print(f"  {key}: {value}")
print()
```

### 1.3.5 Text Similarity (Somiglianza di testi)

**Definizione**: Misurare quanto due testi sono simili.

**Esempi**:
- Matching tra CV e offerta di lavoro
- Duplicate detection (trovare articoli duplicati)
- Raccomandazioni

```python
# --- TASK 5: Text Similarity ---

def text_similarity_overlap(text1, text2):
    """
    Misura di somiglianza semplice: proporzione di parole comuni.
    Approccio naïve: conta le parole che compaiono in entrambi i testi.
    """
    words1 = set(text1.lower().split())
    words2 = set(text2.lower().split())
    
    # Jaccard similarity
    intersection = len(words1 & words2)
    union = len(words1 | words2)
    
    similarity = intersection / union if union > 0 else 0
    return similarity

# Test
cv1 = "Python programming machine learning data analysis"
cv2 = "Python programming data science analytics"
job_posting = "We need Python machine learning data scientist"

print("=== TASK 5: Text Similarity ===")
print(f"CV1: '{cv1}'")
print(f"CV2: '{cv2}'")
print(f"Job: '{job_posting}'")
print()
print(f"Somiglianza CV1-Job: {text_similarity_overlap(cv1, job_posting):.2f}")
print(f"Somiglianza CV2-Job: {text_similarity_overlap(cv2, job_posting):.2f}")
print()
print("Interpretazione: CV1 è più simile all'offerta di lavoro.")
print()
```

### 1.3.6 Machine Translation

**Definizione**: Tradurre testo da una lingua a un'altra.

**Esempi**: Google Translate, traduzioni automatiche nei browser

```python
# --- TASK 6: Machine Translation (semplice, basato su dizionario) ---

def simple_dictionary_translation(text, language='italian_to_english'):
    """
    Traduzione semplice basata su dizionario.
    Approccio naïve: cerca e sostituisci parola per parola.
    (In realtà la traduzione è molto più complessa!)
    """
    # Dizionari semplici
    dictionaries = {
        'italian_to_english': {
            'ciao': 'hello',
            'gatto': 'cat',
            'cane': 'dog',
            'bello': 'beautiful'
        },
        'english_to_italian': {
            'hello': 'ciao',
            'cat': 'gatto',
            'dog': 'cane',
            'beautiful': 'bello'
        }
    }
    
    words = text.lower().split()
    dictionary = dictionaries.get(language, {})
    translated_words = [dictionary.get(word, word) for word in words]
    
    return ' '.join(translated_words)

# Test
print("=== TASK 6: Machine Translation ===")
italian_text = "ciao, gatto bello"
translated = simple_dictionary_translation(italian_text, 'italian_to_english')
print(f"Italiano: '{italian_text}'")
print(f"Inglese: '{translated}'")
print()
print("Nota: Questo è un approccio molto semplice!")
print("La traduzione reale è molto più complessa (dipende dal contesto, grammatica, etc.)")
print()
```

### 1.3.7 Question Answering

**Definizione**: Rispondere a domande su un testo o un argomento.

**Esempi**: FAQ automatiche, assistenti virtuali che rispondono a domande

```python
# --- TASK 7: Question Answering (semplice) ---

def simple_qa_system(question, context):
    """
    Sistema Q&A semplice basato su ricerca di parole chiave.
    Approccio naïve: trova la frase che contiene più parole della domanda.
    """
    # Estrai parole dalla domanda (escludi stop words)
    stop_words = {'what', 'is', 'who', 'where', 'when', 'why', 'how', 'the', 'a'}
    question_words = set(word.lower() for word in question.split() if word.lower() not in stop_words)
    
    # Dividi il contesto in frasi
    sentences = context.split('.')
    
    # Trova la frase che contiene più parole della domanda
    best_sentence = None
    best_score = 0
    
    for sentence in sentences:
        sentence_words = set(word.lower() for word in sentence.split())
        score = len(question_words & sentence_words)
        if score > best_score:
            best_score = score
            best_sentence = sentence.strip()
    
    return best_sentence if best_sentence else "I don't know."

# Test
context = "Milano is a city in Italy. It is known for fashion. The Duomo is a famous cathedral."
questions = [
    "What is Milano?",
    "What is Milano known for?"
]

print("=== TASK 7: Question Answering ===")
print(f"Contesto: '{context}'")
print()
for question in questions:
    answer = simple_qa_system(question, context)
    print(f"Domanda: '{question}'")
    print(f"Risposta: '{answer}'")
    print()
```

### 1.3.8 Text Summarization

**Definizione**: Creare un riassunto breve di un testo più lungo.

**Esempi**: Sommari di articoli, abstract automatici

```python
# --- TASK 8: Text Summarization (semplice) ---

def simple_summarization(text, num_sentences=2):
    """
    Riassunto semplice basato sulla frequenza di parole.
    Approccio naïve: seleziona le frasi con le parole più frequenti.
    """
    from collections import Counter
    
    # Dividi in frasi
    sentences = text.split('.')
    sentences = [s.strip() for s in sentences if s.strip()]
    
    if len(sentences) <= num_sentences:
        return text
    
    # Conta le parole (escludi stop words)
    stop_words = {'the', 'a', 'is', 'was', 'in', 'and', 'of', 'to', 'for'}
    words = text.lower().split()
    words = [w for w in words if w not in stop_words and len(w) > 2]
    word_freq = Counter(words)
    
    # Assegna punteggio a ogni frase
    sentence_scores = {}
    for i, sentence in enumerate(sentences):
        for word in sentence.lower().split():
            if word in word_freq:
                sentence_scores[i] = sentence_scores.get(i, 0) + word_freq[word]
    
    # Seleziona le top-N frasi
    top_sentences = sorted(sentence_scores.items(), key=lambda x: x[1], reverse=True)[:num_sentences]
    top_sentences = sorted(top_sentences, key=lambda x: x[0])  # Ordina per indice originale
    
    summary = '. '.join([sentences[i] for i, _ in top_sentences])
    return summary + '.'

# Test
article = "Machine learning is a subset of artificial intelligence. It enables computers to learn from data. Deep learning uses neural networks. These techniques are transforming many industries."

print("=== TASK 8: Text Summarization ===")
print(f"Testo originale: '{article}'")
print()
summary = simple_summarization(article, num_sentences=2)
print(f"Riassunto: '{summary}'")
print()
```

### 1.3.9 Text Generation

**Definizione**: Generare testo nuovo, basato su pattern appresi dai dati.

**Esempi**: Completamento automatico, generazione di risposte, chatbot

```python
# --- TASK 9: Text Generation (semplice) ---

def simple_text_completion(initial_text, knowledge_base):
    """
    Completamento di testo semplice basato su una knowledge base.
    Approccio naïve: continua con frasi comuni dopo la parola finale.
    """
    # Knowledge base: mapping da parola -> parole che seguono frequentemente
    # (In realtà questo si impara da dati enormi)
    word_continuations = {
        'machine': ['learning', 'learning is', 'learning models'],
        'data': ['science', 'scientist', 'analysis'],
        'learning': ['is fun', 'is useful', 'algorithms'],
    }
    
    last_word = initial_text.lower().split()[-1]
    
    if last_word in word_continuations:
        # Prendi la continuazione più probabile
        continuation = word_continuations[last_word][0]
        return initial_text + ' ' + continuation
    else:
        return initial_text + ' [unknown next word]'

# Test
print("=== TASK 9: Text Generation ===")
test_texts = [
    "Machine learning",
    "Data science",
    "Learning algorithms"
]
knowledge_base = {}

for text in test_texts:
    completed = simple_text_completion(text, knowledge_base)
    print(f"Input: '{text}'")
    print(f"Completato: '{completed}'")
    print()
```
