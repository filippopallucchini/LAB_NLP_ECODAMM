## 1.2 L'obiettivo: rappresentare il testo

### La sfida centrale di NLP

Il compito principale dell'NLP è trovare una **rappresentazione numerica efficace** di parole, frasi e testi.

Perché? Perché:
1. I computer capiscono i **numeri**, non le parole
2. Una volta che il testo è numerico, possiamo applicare algoritmi di Machine Learning
3. Diverse rappresentazioni catturano diverse proprietà del testo

### Confronto: Testo grezzo vs. Rappresentazione numerica

```python
import numpy as np

# Testo grezzo (quello che gli umani leggono)
sentence = "I love machine learning"
print("=== TESTO GREZZO ===")
print(f"Frase: {sentence}")
print(f"Tipo: {type(sentence)}")
print()

# Rappresentazione 1: Semplice - codificare ogni parola con un numero
print("=== RAPPRESENTAZIONE 1: One-Hot Encoding (Simbolico) ===")
vocab = {"I": 0, "love": 1, "machine": 2, "learning": 3}
words = sentence.split()
encoded = [vocab[word] for word in words]
print(f"Parole: {words}")
print(f"Codificate: {encoded}")
print(f"Tipo: lista di interi")
print()

# Rappresentazione 2: Bag of Words (contare le parole)
print("=== RAPPRESENTAZIONE 2: Bag of Words (Statistica) ===")
from collections import Counter
word_counts = Counter(words)
bow_vector = [word_counts.get(word, 0) for word in sorted(vocab.keys())]
print(f"Vocabolario ordinato: {sorted(vocab.keys())}")
print(f"Bag of Words: {bow_vector}")
print(f"Significato: [1 volta 'I', 1 volta 'love', 1 volta 'machine', 1 volta 'learning']")
print()

# Rappresentazione 3: Embedding (vettore denso, simula i significati)
print("=== RAPPRESENTAZIONE 3: Word Embeddings (Deep Learning) ===")
np.random.seed(42)
# In realtà gli embeddings si calcolano da dati enormi, qui usiamo numeri casuali per illustrare
embeddings = {
    "I": np.array([0.2, -0.5, 0.8]),
    "love": np.array([0.9, 0.3, -0.1]),
    "machine": np.array([0.1, 0.7, 0.4]),
    "learning": np.array([0.3, 0.6, 0.5])
}
print(f"Ogni parola è un vettore denso (3 dimensioni in questo esempio):")
for word in words:
    print(f"  {word}: {embeddings[word]}")
print()
# Calcola il vettore medio della frase
sentence_embedding = np.mean([embeddings[word] for word in words], axis=0)
print(f"Embedding della FRASE (media): {sentence_embedding}")
print(f"Idea: parole simili hanno vettori simili")
```

### 1.3.1 Text Classification (Classificazione di testi)

**Definizione**: Assegnare a un testo una o più categorie predefinite.

**Esempi**:
- **Spam detection**: Un'email è spam o no?
- **Sentiment analysis**: Una recensione è positiva, negativa o neutra?
- **Topic classification**: Un articolo parla di sport, politica, tecnologia o economia?

```python
# --- TASK 1: Text Classification (Sentiment Analysis) ---

def simple_sentiment_classifier(text):
    """
    Classificatore basato su dizionario (Symbolic approach).
    Approccio naïve: conta le parole positive e negative.
    """
    positive_words = ['good', 'great', 'excellent', 'wonderful', 'amazing', 'love', 'beautiful']
    negative_words = ['bad', 'terrible', 'awful', 'horrible', 'hate', 'ugly', 'disgusting']
    
    text_lower = text.lower()
    positive_count = sum(1 for word in positive_words if word in text_lower)
    negative_count = sum(1 for word in negative_words if word in text_lower)
    
    if positive_count > negative_count:
        return "POSITIVE", positive_count - negative_count
    elif negative_count > positive_count:
        return "NEGATIVE", negative_count - positive_count
    else:
        return "NEUTRAL", 0

# Test
reviews = [
    "This movie is absolutely wonderful and amazing!",
    "Terrible and awful experience, I hate it.",
    "The weather is nice today."
]

print("=== TASK 1: Text Classification (Sentiment Analysis) ===")
for review in reviews:
    sentiment, score = simple_sentiment_classifier(review)
    print(f"Testo: '{review}'")
    print(f"Sentimento: {sentiment} (score: {score})")
    print()
```

### 1.3.2 Named Entity Recognition (NER)

**Definizione**: Identificare e classificare entità nominate (persone, luoghi, organizzazioni, ecc.) nel testo.

**Esempi**:
- Trovare nomi di persone: "Mario Rossi"
- Trovare nomi di città: "Milano", "Roma"
- Trovare nomi di aziende: "Apple", "Microsoft"

```python
import re

# --- TASK 2: Named Entity Recognition (NER) ---

def simple_ner(text):
    """
    Riconoscimento semplice di entità basato su regole e dizionari.
    Approccio naïve: usa una lista di entità conosciute.
    """
    # Dizionari di entità conosciute
    people = ['Mario', 'Luigi', 'John', 'Alice']
    cities = ['Milano', 'Roma', 'Venice', 'New York']
    companies = ['Apple', 'Google', 'Microsoft', 'Amazon']
    
    entities = {}
    
    # Cerca persone
    for person in people:
        if person in text:
            entities[person] = 'PERSON'
    
    # Cerca città
    for city in cities:
        if city in text:
            entities[city] = 'LOCATION'
    
    # Cerca aziende
    for company in companies:
        if company in text:
            entities[company] = 'ORGANIZATION'
    
    return entities

# Test
texts = [
    "Mario lives in Milano and works for Google.",
    "Alice visited Rome and had lunch at an Apple store."
]

print("=== TASK 2: Named Entity Recognition (NER) ===")
for text in texts:
    entities = simple_ner(text)
    print(f"Testo: '{text}'")
    print(f"Entità trovate: {entities}")
    print()
```

### 1.3.3 Part-of-Speech (POS) Tagging

**Definizione**: Assegnare a ogni parola una categoria grammaticale (sostantivo, verbo, aggettivo, ecc.).

**Esempi**:
- "The cat is sleeping" → The (DET), cat (NOUN), is (VERB), sleeping (VERB)

```python
# --- TASK 3: Part-of-Speech Tagging ---

def simple_pos_tagger(sentence):
    """
    Etichettatura semplice basata su dizionario.
    Approccio naïve: assegna la categoria più comune.
    """
    # Dizionario parola -> categoria
    pos_dict = {
        'the': 'DET',
        'a': 'DET',
        'cat': 'NOUN',
        'dog': 'NOUN',
        'is': 'VERB',
        'runs': 'VERB',
        'beautiful': 'ADJ',
        'happy': 'ADJ',
        'quickly': 'ADV',
        'slowly': 'ADV'
    }
    
    words = sentence.lower().split()
    tagged = []
    
    for word in words:
        # Rimuovi punteggiatura
        clean_word = word.strip('.,!?;:')
        tag = pos_dict.get(clean_word, 'UNKNOWN')
        tagged.append((word, tag))
    
    return tagged

# Test
sentences = [
    "The cat is beautiful.",
    "A dog runs quickly."
]

print("=== TASK 3: Part-of-Speech Tagging ===")
for sentence in sentences:
    tagged = simple_pos_tagger(sentence)
    print(f"Frase: '{sentence}'")
    print("POS tags:")
    for word, tag in tagged:
        print(f"  {word}: {tag}")
    print()
```

### 1.3.4 Information Extraction (IE)

**Definizione**: Estrarre informazioni strutturate (relazioni, fatti) da testi non strutturati.

**Esempio**: Da un contratto, estrarre: data_firma, partecipanti, importo, condizioni.

```python
# --- TASK 4: Information Extraction ---

def simple_information_extraction(text):
    """
    Estrazione di informazioni basata su pattern e regex.
    Approccio naïve: cerca parole chiave e pattern specifici.
    """
    extracted_info = {}
    
    # Cerca importo (numero seguito da $, €, ecc.)
    amount_match = re.search(r'\$(\d+)', text)
    if amount_match:
        extracted_info['amount'] = amount_match.group(1)
    
    # Cerca date (formato DD/MM/YYYY)
    date_match = re.search(r'(\d{1,2}/\d{1,2}/\d{4})', text)
    if date_match:
        extracted_info['date'] = date_match.group(1)
    
    # Cerca nomi (parole capitalizzate)
    names = re.findall(r'\b([A-Z][a-z]+)\b', text)
    if names:
        extracted_info['names'] = names
    
    return extracted_info

# Test
contract = "Agreement between Mario and Alice on 15/03/2024 for $5000."

print("=== TASK 4: Information Extraction ===")
print(f"Testo: '{contract}'")
info = simple_information_extraction(contract)
print(f"Informazioni estratte:")
for key, value in info.items():
    print(f"  {key}: {value}")
print()
```

### 1.3.5 Text Similarity (Somiglianza di testi)

**Definizione**: Misurare quanto due testi sono simili.

**Esempi**:
- Matching tra CV e offerta di lavoro
- Duplicate detection (trovare articoli duplicati)
- Raccomandazioni

```python
# --- TASK 5: Text Similarity ---

def text_similarity_overlap(text1, text2):
    """
    Misura di somiglianza semplice: proporzione di parole comuni.
    Approccio naïve: conta le parole che compaiono in entrambi i testi.
    """
    words1 = set(text1.lower().split())
    words2 = set(text2.lower().split())
    
    # Jaccard similarity
    intersection = len(words1 & words2)
    union = len(words1 | words2)
    
    similarity = intersection / union if union > 0 else 0
    return similarity

# Test
cv1 = "Python programming machine learning data analysis"
cv2 = "Python programming data science analytics"
job_posting = "We need Python machine learning data scientist"

print("=== TASK 5: Text Similarity ===")
print(f"CV1: '{cv1}'")
print(f"CV2: '{cv2}'")
print(f"Job: '{job_posting}'")
print()
print(f"Somiglianza CV1-Job: {text_similarity_overlap(cv1, job_posting):.2f}")
print(f"Somiglianza CV2-Job: {text_similarity_overlap(cv2, job_posting):.2f}")
print()
print("Interpretazione: CV1 è più simile all'offerta di lavoro.")
print()
```

### 1.3.6 Machine Translation

**Definizione**: Tradurre testo da una lingua a un'altra.

**Esempi**: Google Translate, traduzioni automatiche nei browser

```python
# --- TASK 6: Machine Translation (semplice, basato su dizionario) ---

def simple_dictionary_translation(text, language='italian_to_english'):
    """
    Traduzione semplice basata su dizionario.
    Approccio naïve: cerca e sostituisci parola per parola.
    (In realtà la traduzione è molto più complessa!)
    """
    # Dizionari semplici
    dictionaries = {
        'italian_to_english': {
            'ciao': 'hello',
            'gatto': 'cat',
            'cane': 'dog',
            'bello': 'beautiful'
        },
        'english_to_italian': {
            'hello': 'ciao',
            'cat': 'gatto',
            'dog': 'cane',
            'beautiful': 'bello'
        }
    }
    
    words = text.lower().split()
    dictionary = dictionaries.get(language, {})
    translated_words = [dictionary.get(word, word) for word in words]
    
    return ' '.join(translated_words)

# Test
print("=== TASK 6: Machine Translation ===")
italian_text = "ciao, gatto bello"
translated = simple_dictionary_translation(italian_text, 'italian_to_english')
print(f"Italiano: '{italian_text}'")
print(f"Inglese: '{translated}'")
print()
print("Nota: Questo è un approccio molto semplice!")
print("La traduzione reale è molto più complessa (dipende dal contesto, grammatica, etc.)")
print()
```

### 1.3.7 Question Answering

**Definizione**: Rispondere a domande su un testo o un argomento.

**Esempi**: FAQ automatiche, assistenti virtuali che rispondono a domande

```python
# --- TASK 7: Question Answering (semplice) ---

def simple_qa_system(question, context):
    """
    Sistema Q&A semplice basato su ricerca di parole chiave.
    Approccio naïve: trova la frase che contiene più parole della domanda.
    """
    # Estrai parole dalla domanda (escludi stop words)
    stop_words = {'what', 'is', 'who', 'where', 'when', 'why', 'how', 'the', 'a'}
    question_words = set(word.lower() for word in question.split() if word.lower() not in stop_words)
    
    # Dividi il contesto in frasi
    sentences = context.split('.')
    
    # Trova la frase che contiene più parole della domanda
    best_sentence = None
    best_score = 0
    
    for sentence in sentences:
        sentence_words = set(word.lower() for word in sentence.split())
        score = len(question_words & sentence_words)
        if score > best_score:
            best_score = score
            best_sentence = sentence.strip()
    
    return best_sentence if best_sentence else "I don't know."

# Test
context = "Milano is a city in Italy. It is known for fashion. The Duomo is a famous cathedral."
questions = [
    "What is Milano?",
    "What is Milano known for?"
]

print("=== TASK 7: Question Answering ===")
print(f"Contesto: '{context}'")
print()
for question in questions:
    answer = simple_qa_system(question, context)
    print(f"Domanda: '{question}'")
    print(f"Risposta: '{answer}'")
    print()
```

### 1.3.8 Text Summarization

**Definizione**: Creare un riassunto breve di un testo più lungo.

**Esempi**: Sommari di articoli, abstract automatici

```python
# --- TASK 8: Text Summarization (semplice) ---

def simple_summarization(text, num_sentences=2):
    """
    Riassunto semplice basato sulla frequenza di parole.
    Approccio naïve: seleziona le frasi con le parole più frequenti.
    """
    from collections import Counter
    
    # Dividi in frasi
    sentences = text.split('.')
    sentences = [s.strip() for s in sentences if s.strip()]
    
    if len(sentences) <= num_sentences:
        return text
    
    # Conta le parole (escludi stop words)
    stop_words = {'the', 'a', 'is', 'was', 'in', 'and', 'of', 'to', 'for'}
    words = text.lower().split()
    words = [w for w in words if w not in stop_words and len(w) > 2]
    word_freq = Counter(words)
    
    # Assegna punteggio a ogni frase
    sentence_scores = {}
    for i, sentence in enumerate(sentences):
        for word in sentence.lower().split():
            if word in word_freq:
                sentence_scores[i] = sentence_scores.get(i, 0) + word_freq[word]
    
    # Seleziona le top-N frasi
    top_sentences = sorted(sentence_scores.items(), key=lambda x: x[1], reverse=True)[:num_sentences]
    top_sentences = sorted(top_sentences, key=lambda x: x[0])  # Ordina per indice originale
    
    summary = '. '.join([sentences[i] for i, _ in top_sentences])
    return summary + '.'

# Test
article = "Machine learning is a subset of artificial intelligence. It enables computers to learn from data. Deep learning uses neural networks. These techniques are transforming many industries."

print("=== TASK 8: Text Summarization ===")
print(f"Testo originale: '{article}'")
print()
summary = simple_summarization(article, num_sentences=2)
print(f"Riassunto: '{summary}'")
print()
```

### 1.3.9 Text Generation

**Definizione**: Generare testo nuovo, basato su pattern appresi dai dati.

**Esempi**: Completamento automatico, generazione di risposte, chatbot

```python
# --- TASK 9: Text Generation (semplice) ---

def simple_text_completion(initial_text, knowledge_base):
    """
    Completamento di testo semplice basato su una knowledge base.
    Approccio naïve: continua con frasi comuni dopo la parola finale.
    """
    # Knowledge base: mapping da parola -> parole che seguono frequentemente
    # (In realtà questo si impara da dati enormi)
    word_continuations = {
        'machine': ['learning', 'learning is', 'learning models'],
        'data': ['science', 'scientist', 'analysis'],
        'learning': ['is fun', 'is useful', 'algorithms'],
    }
    
    last_word = initial_text.lower().split()[-1]
    
    if last_word in word_continuations:
        # Prendi la continuazione più probabile
        continuation = word_continuations[last_word][0]
        return initial_text + ' ' + continuation
    else:
        return initial_text + ' [unknown next word]'

# Test
print("=== TASK 9: Text Generation ===")
test_texts = [
    "Machine learning",
    "Data science",
    "Learning algorithms"
]
knowledge_base = {}

for text in test_texts:
    completed = simple_text_completion(text, knowledge_base)
    print(f"Input: '{text}'")
    print(f"Completato: '{completed}'")
    print()
```

## 1.4 I tre approcci principali dell'NLP

Nei prossimi notebook approfondiremo tre approcci diversi per affrontare i task dell'NLP:

### 1. **Approccio Simbolico (Symbolic/Rule-based)**
- **Idea**: Usa regole, dizionari e logica esplicita
- **Strumenti**: WordNet, VADER (Valence Aware Dictionary and sEntiment Reasoner), regex
- **Vantaggi**: Interpretabile, controllabile, rapido
- **Svantaggi**: Laborioso da costruire, difficile a generalizzare

### 2. **Approccio Bag of Words (Statistica)**
- **Idea**: Rappresenta il testo come frequenza di parole (ignora l'ordine)
- **Strumenti**: TF-IDF, Count Vectorizer, Naive Bayes
- **Vantaggi**: Semplice, veloce, funziona bene per molti task
- **Svantaggi**: Perde l'informazione sull'ordine e il contesto

### 3. **Approccio Word Embeddings (Deep Learning)**
- **Idea**: Rappresenta le parole come vettori densi che catturano significato e relazioni
- **Strumenti**: Word2Vec, GloVe, FastText, BERT, GPT
- **Vantaggi**: Cattura relazioni semantiche, stato dell'arte per molti task
- **Svantaggi**: Richiede molti dati, meno interpretabile

### Confronto pratico: Tre approcci sulla stessa frase

```python
print("="*80)
print("CONFRONTO: I TRE APPROCCI DELL'NLP")
print("="*80)
print()

sentence = "I love machine learning"
print(f"Frase di esempio: '{sentence}'")
print()

# APPROCCIO 1: SIMBOLICO (Rule-based)
print("-" * 80)
print("APPROCCIO 1: SIMBOLICO (Rule-based / Lexicon-based)")
print("-" * 80)
print()
print("Idea: Uso di regole esplicite e dizionari")
print()
positive_words = ['love', 'excellent', 'amazing']
negative_words = ['hate', 'terrible', 'awful']

words_in_sentence = sentence.lower().split()
positive_found = [w for w in words_in_sentence if w in positive_words]
negative_found = [w for w in words_in_sentence if w in negative_words]

print(f"Parole positive trovate: {positive_found}")
print(f"Parole negative trovate: {negative_found}")
if len(positive_found) > len(negative_found):
    print(f"Conclusione (basata su regole): POSITIVO (Sentiment)")
else:
    print(f"Conclusione (basata su regole): NEGATIVO (Sentiment)")
print()
print("Vantaggi: Controllabile, veloce, interpretabile")
print("Svantaggi: Rigido, non generalizza bene a nuovi contesti")
print()

# APPROCCIO 2: BAG OF WORDS (Statistica)
print("-" * 80)
print("APPROCCIO 2: BAG OF WORDS (Statistica)")
print("-" * 80)
print()
print("Idea: Rappresenta il testo come conteggio (frequenza) delle parole")
print()

# Crea un vocabolario
vocab = ['I', 'love', 'machine', 'learning', 'hate', 'hate', 'deep']  # ordered vocab
vocab_unique = sorted(set(vocab))
print(f"Vocabolario: {vocab_unique}")
print()

# Crea il bag of words
bow = []
for word in vocab_unique:
    count = sentence.lower().split().count(word.lower())
    bow.append(count)

print(f"Bag of Words: {bow}")
print(f"Significato: [I:{bow[0]}, deep:{bow[1]}, hate:{bow[2]}, learning:{bow[3]}, love:{bow[4]}, machine:{bow[5]}]")
print()
print("Nota: L'ordine originale ('I love machine learning') è PERSO!")
print("Il BoW vede 'learning love machine I' come identico.")
print()
print("Vantaggi: Semplice, veloce, funziona bene per molti task")
print("Svantaggi: Perde l'ordine e il contesto, non cattura relazioni semantiche")
print()

# APPROCCIO 3: WORD EMBEDDINGS (Deep Learning)
print("-" * 80)
print("APPROCCIO 3: WORD EMBEDDINGS (Deep Learning)")
print("-" * 80)
print()
print("Idea: Rappresenta ogni parola come vettore denso che cattura significato e relazioni")
print()

np.random.seed(42)
word_embeddings = {
    'I': np.array([0.1, 0.2, -0.3]),
    'love': np.array([0.9, 0.8, 0.2]),
    'machine': np.array([0.6, -0.1, 0.7]),
    'learning': np.array([0.5, 0.0, 0.6])
}

print("Embeddings delle parole (3 dimensioni):")
for word, embedding in word_embeddings.items():
    print(f"  {word}: {embedding}")
print()

# Calcola somiglianze
print("Somiglianze tra vettori (prodotto scalare normalizzato):")
from numpy.linalg import norm

def cosine_similarity(v1, v2):
    return np.dot(v1, v2) / (norm(v1) * norm(v2))

sim_love_learning = cosine_similarity(word_embeddings['love'], word_embeddings['learning'])
sim_love_machine = cosine_similarity(word_embeddings['love'], word_embeddings['machine'])
print(f"  Somiglianza 'love' - 'learning': {sim_love_learning:.3f}")
print(f"  Somiglianza 'love' - 'machine': {sim_love_machine:.3f}")
print()

# Rappresentazione della frase
sentence_embedding = np.mean([word_embeddings[w.lower()] for w in words_in_sentence], axis=0)
print(f"Embedding della FRASE (media dei word embeddings): {sentence_embedding}")
print()
print("Idea: Parole simili hanno vettori simili!")
print("  - 'love' e 'learning' sono simili (entrambi positivi nel contesto)")
print()
print("Vantaggi: Cattura significato e relazioni, stato dell'arte, generalizza bene")
print("Svantaggi: Richiede molti dati, meno interpretabile, computazionalmente costoso")
print()
```

### PROBLEMA 1: LinkedIn Job Matcher

**Task**: Text Similarity

**Soluzione**: Calcolare la sovrapposizione di skill tra CV e job posting

```python
# SOLUZIONE 1: Linkedin Job Matcher (Text Similarity)

def linkedin_job_matcher(cv_text, job_posting):
    """
    Matcher basato su somiglianza di testo (Jaccard similarity).
    Idea: Conta quante skill nel CV sono richieste nel job posting.
    """
    # Converti in lowercase e splitta
    cv_skills = set(cv_text.lower().split())
    job_skills = set(job_posting.lower().split())
    
    # Calcola Jaccard similarity
    intersection = cv_skills & job_skills
    union = cv_skills | job_skills
    
    similarity = len(intersection) / len(union) if len(union) > 0 else 0
    
    # Matchate skills
    matched_skills = intersection
    
    return {
        'similarity_score': similarity,
        'matched_skills': matched_skills,
        'missing_skills': job_skills - cv_skills
    }

print("="*80)
print("SOLUZIONE 1: LinkedIn Job Matcher (Text Similarity)")
print("="*80)
print()

# Test 1
print("Test 1:")
cv1 = "Python Machine Learning Data Analysis SQL"
job1 = "Python Data Science Machine Learning Statistics"
result1 = linkedin_job_matcher(cv1, job1)
print(f"CV: {cv1}")
print(f"Job: {job1}")
print(f"Match Score: {result1['similarity_score']:.2%}")
print(f"Matched Skills: {result1['matched_skills']}")
print(f"Missing Skills: {result1['missing_skills']}")
print()

# Test 2
print("Test 2:")
cv2 = "Java C++ Web Development HTML CSS"
job2 = "Python Machine Learning Data Science"
result2 = linkedin_job_matcher(cv2, job2)
print(f"CV: {cv2}")
print(f"Job: {job2}")
print(f"Match Score: {result2['similarity_score']:.2%}")
print(f"Matched Skills: {result2['matched_skills']}")
print(f"Missing Skills: {result2['missing_skills']}")
print()
print()
```

### PROBLEMA 2: Email Filter

**Task**: Text Classification (Spam Detection)

**Soluzione**: Classificare email basato su parole spam comuni

```python
# SOLUZIONE 2: Email Filter (Text Classification - Spam Detection)

def email_spam_filter(email_text):
    """
    Filtro spam basato su dizionario di parole spam.
    Idea: Conta le parole spam e classifica come spam se supera una soglia.
    """
    spam_keywords = [
        'click here', 'buy now', 'free', 'winner', 'congratulations',
        'limited time', 'act now', 'urgent', 'verify account', 'confirm password',
        'nigerian', 'inheritance', 'bank transfer', 'casino', 'lottery'
    ]
    
    email_lower = email_text.lower()
    spam_count = sum(1 for keyword in spam_keywords if keyword in email_lower)
    
    # Soglia: se contiene più di 1 spam keyword, è spam
    is_spam = spam_count > 1
    spam_probability = min(spam_count / 3, 1.0)  # Normalizza a [0, 1]
    
    return {
        'classification': 'SPAM' if is_spam else 'LEGITIMATE',
        'spam_probability': spam_probability,
        'spam_keywords_found': spam_count
    }

print("="*80)
print("SOLUZIONE 2: Email Filter (Text Classification - Spam Detection)")
print("="*80)
print()

# Test 1 (Spam)
print("Test 1 (Spam):")
email1 = "Congratulations! You are a winner! Click here now to claim your free prize!"
result1 = email_spam_filter(email1)
print(f"Email: {email1}")
print(f"Classification: {result1['classification']}")
print(f"Spam Probability: {result1['spam_probability']:.2%}")
print(f"Spam Keywords Found: {result1['spam_keywords_found']}")
print()

# Test 2 (Legitimate)
print("Test 2 (Legitimate):")
email2 = "Hi Alice, let's schedule a meeting for next week. Best regards, Bob."
result2 = email_spam_filter(email2)
print(f"Email: {email2}")
print(f"Classification: {result2['classification']}")
print(f"Spam Probability: {result2['spam_probability']:.2%}")
print(f"Spam Keywords Found: {result2['spam_keywords_found']}")
print()
print()
```

### PROBLEMA 3: Financial Report Analyzer

**Task**: Information Extraction

**Soluzione**: Estrarre importi, date e nomi usando regex

```python
# SOLUZIONE 3: Financial Report Analyzer (Information Extraction)

def financial_report_analyzer(report_text):
    """
    Estrae informazioni finanziarie da un rapporto.
    Usa regex per trovare: importi, date, nomi di aziende.
    """
    extracted = {
        'amounts': [],
        'dates': [],
        'companies': []
    }
    
    # Cerca importi (numero seguiti da $ o euro)
    amount_pattern = r'\$([\d,]+)|€([\d,]+)|([\d,]+)\s*(dollars?|euros?|million|billion)'
    amounts = re.findall(amount_pattern, report_text, re.IGNORECASE)
    extracted['amounts'] = [' '.join(filter(None, m)) for m in amounts]
    
    # Cerca date (format DD/MM/YYYY o MM-DD-YYYY o YYYY-MM-DD)
    date_pattern = r'\d{1,2}[-/]\d{1,2}[-/]\d{4}|\d{4}[-/]\d{1,2}[-/]\d{1,2}'
    dates = re.findall(date_pattern, report_text)
    extracted['dates'] = dates
    
    # Cerca nomi di aziende (parole capitalizzate)
    company_pattern = r'\b([A-Z][a-z]+(?:\s+[A-Z][a-z]+)*)\b'
    companies = re.findall(company_pattern, report_text)
    # Filtra stop words
    stop_words = {'The', 'A', 'An', 'In', 'Of', 'And', 'Or', 'To', 'For'}
    extracted['companies'] = [c for c in companies if c not in stop_words]
    
    return extracted

print("="*80)
print("SOLUZIONE 3: Financial Report Analyzer (Information Extraction)")
print("="*80)
print()

# Test 1
print("Test 1:")
report1 = "Apple Inc. reported revenue of $365 billion for Q1 2024. Date: 01/31/2024. Microsoft earned €150 million."
result1 = financial_report_analyzer(report1)
print(f"Report: {report1}")
print(f"Amounts Found: {result1['amounts']}")
print(f"Dates Found: {result1['dates']}")
print(f"Companies Found: {result1['companies']}")
print()

# Test 2
print("Test 2:")
report2 = "Google and Amazon signed a contract worth $50 million on 2024-03-15."
result2 = financial_report_analyzer(report2)
print(f"Report: {report2}")
print(f"Amounts Found: {result2['amounts']}")
print(f"Dates Found: {result2['dates']}")
print(f"Companies Found: {result2['companies']}")
print()
print()
```

### PROBLEMA 4: FAQ Chatbot

**Task**: Question Answering

**Soluzione**: Rispondere usando ricerca di somiglianza di domande

```python
# SOLUZIONE 4: FAQ Chatbot (Question Answering)

def faq_chatbot(user_question, faq_database):
    """
    Chatbot FAQ basato su matching di somiglianza.
    Idea: Trova la domanda FAQ più simile a quella dell'utente.
    
    Args:
        user_question: La domanda dell'utente
        faq_database: Dizionario {domanda_faq: risposta}
    """
    def word_overlap_similarity(q1, q2):
        """Calcola somiglianza come overlap di parole."""
        words1 = set(q1.lower().split())
        words2 = set(q2.lower().split())
        intersection = words1 & words2
        union = words1 | words2
        return len(intersection) / len(union) if len(union) > 0 else 0
    
    best_match = None
    best_score = 0
    
    # Trova la domanda FAQ più simile
    for faq_question, answer in faq_database.items():
        score = word_overlap_similarity(user_question, faq_question)
        if score > best_score:
            best_score = score
            best_match = answer
    
    confidence = "HIGH" if best_score > 0.5 else "MEDIUM" if best_score > 0.3 else "LOW"
    
    return {
        'answer': best_match if best_match else "I don't know. Please contact support.",
        'confidence': confidence,
        'match_score': best_score
    }

print("="*80)
print("SOLUZIONE 4: FAQ Chatbot (Question Answering)")
print("="*80)
print()

# FAQ database
faq_db = {
    "What are your business hours?": "We are open Monday-Friday 9AM-6PM, Saturday 10AM-4PM.",
    "How do I reset my password?": "Click on 'Forgot Password' on the login page and follow the steps.",
    "What is your return policy?": "We accept returns within 30 days of purchase with original receipt.",
    "Do you offer free shipping?": "Free shipping for orders over $50. Standard shipping costs $5.99."
}

# Test 1
print("Test 1:")
user_q1 = "When are you guys open?"
result1 = faq_chatbot(user_q1, faq_db)
print(f"User: '{user_q1}'")
print(f"Bot Answer: {result1['answer']}")
print(f"Confidence: {result1['confidence']} (score: {result1['match_score']:.2f})")
print()

# Test 2
print("Test 2:")
user_q2 = "Can I return items if I'm not satisfied?"
result2 = faq_chatbot(user_q2, faq_db)
print(f"User: '{user_q2}'")
print(f"Bot Answer: {result2['answer']}")
print(f"Confidence: {result2['confidence']} (score: {result2['match_score']:.2f})")
print()
print()
```

### PROBLEMA 5: Language Translator for Social Media

**Task**: Machine Translation

**Soluzione**: Traduzione basata su dizionari

```python
# SOLUZIONE 5: Language Translator for Social Media (Machine Translation)

def social_media_translator(text, source_lang, target_lang, translation_dict):
    """
    Traduttore semplice basato su dizionario.
    Nota: Questo è un approccio molto basico! La traduzione reale è molto più complessa.
    
    Args:
        text: Il testo da tradurre
        source_lang: Lingua sorgente (es. 'italian')
        target_lang: Lingua target (es. 'english')
        translation_dict: Dizionario {source_lang -> {parola: traduzione}}
    """
    # Ottieni il dizionario di traduzione
    if source_lang not in translation_dict:
        return {"error": f"Language pair {source_lang}->{target_lang} not supported"}
    
    lang_dict = translation_dict[source_lang].get(target_lang, {})
    
    # Traduce parola per parola
    words = text.split()
    translated_words = []
    unknown_words = []
    
    for word in words:
        # Preserva punteggiatura
        clean_word = word.rstrip('.,!?;:').lower()
        punctuation = word[len(clean_word):]
        
        if clean_word in lang_dict:
            translated_words.append(lang_dict[clean_word] + punctuation)
        else:
            # Se la parola non è nel dizionario, mantienila come è
            translated_words.append(word)
            unknown_words.append(clean_word)
    
    translated_text = ' '.join(translated_words)
    
    return {
        'original_text': text,
        'translated_text': translated_text,
        'unknown_words': unknown_words,
        'coverage': f"{((len(words) - len(unknown_words)) / len(words) * 100):.1f}%"
    }

print("="*80)
print("SOLUZIONE 5: Language Translator for Social Media (Machine Translation)")
print("="*80)
print()

# Dizionari di traduzione
translation_database = {
    'italian': {
        'english': {
            'ciao': 'hello',
            'mondo': 'world',
            'come': 'how',
            'stai': 'are you',
            'bene': 'fine',
            'amore': 'love',
            'gratzie': 'thanks',
            'scusa': 'sorry',
            'mi': 'me',
            'piace': 'like',
            'questo': 'this',
            'bellissimo': 'beautiful',
            'pizza': 'pizza',
            'pasta': 'pasta'
        }
    },
    'spanish': {
        'english': {
            'hola': 'hello',
            'mundo': 'world',
            'gracias': 'thank you',
            'por': 'for',
            'favor': 'please',
            'me': 'me',
            'gusta': 'like',
            'hermoso': 'beautiful'
        }
    }
}

# Test 1 (Italian to English)
print("Test 1 (Italian to English):")
text1 = "Ciao mondo! Come stai?"
result1 = social_media_translator(text1, 'italian', 'english', translation_database)
print(f"Original (Italian): {result1['original_text']}")
print(f"Translated (English): {result1['translated_text']}")
print(f"Coverage: {result1['coverage']}")
if result1['unknown_words']:
    print(f"Unknown words: {result1['unknown_words']}")
print()

# Test 2 (Spanish to English)
print("Test 2 (Spanish to English):")
text2 = "Hola amigos! Gracias por favor!"
result2 = social_media_translator(text2, 'spanish', 'english', translation_database)
print(f"Original (Spanish): {result2['original_text']}")
print(f"Translated (English): {result2['translated_text']}")
print(f"Coverage: {result2['coverage']}")
if result2['unknown_words']:
    print(f"Unknown words: {result2['unknown_words']}")
print()
print()
```
