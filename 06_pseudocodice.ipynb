# 🧠 Lo Pseudocodice: Pensare prima di Programmare

## Benvenuti! Prima di scrivere anche una sola riga di codice in Python, è fondamentale imparare a progettare la logica del nostro programma. Per farlo, si usa lo Pseudocodice.

**Che cos'è?**
**Lo pseudocodice è un linguaggio intermedio tra l'italiano e la programmazione. Non ha regole rigide (niente errori di sintassi!), ma serve a descrivere i passi che un computer deve compiere per risolvere un problema.**

I vantaggi:

- Ti permette di concentrarti sulla soluzione, non sulle parentesi.

- È universale: chiunque può capirlo.

- Riduce drasticamente gli errori quando scriverai il codice vero.

## Esempio 1: Logica Semplice (Trovare il numero massimo)

### Immaginiamo di dover trovare il numero più grande in una lista di voti.

```python
INIZIO
    LEGGI lista_voti
    IMPOSTA massimo = primo elemento di lista_voti
    
    PER OGNI voto IN lista_voti:
        SE voto è maggiore di massimo:
            massimo = voto
            
    STAMPA "Il voto più alto è: " + massimo
FINE
```

```python
sst2 = load_dataset("stanfordnlp/sst2", split="validation")

df_sst = pd.DataFrame(sst2)

# 2. Campionamento Bilanciato
n_campioni = df_sst['label'].value_counts().min()
df_balanced = df_sst.groupby('label').sample(n=n_campioni, random_state=42)

print(f"Dataset bilanciato: {n_campioni} esempi positivi e {n_campioni} negativi.")
generic_pipe = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

texts = df_balanced['sentence'].tolist()
y_true = df_balanced['label'].tolist()

print("Analisi in corso... DistilBERT sta elaborando i testi su CPU.")
results = generic_pipe(texts)

# 5. Mapping e Calcolo Accuratezza
# Il modello restituisce 'POSITIVE' e 'NEGATIVE'. Mappiamo per confrontare con le label del dataset (0, 1)
mapping = {'NEGATIVE': 0, 'POSITIVE': 1}
y_pred = [mapping[r['label']] for r in results]

print(f"Accuratezza Modello DistilBERT su SST-2: {accuracy_score(y_true, y_pred):.2f}")

# 6. Vediamo un esempio pratico dai risultati
for i in range(3):
    print(f"\nTesto: {texts[i]}")
    print(f"Reale: {y_true[i]} | Predetto: {y_pred[i]}")
```
